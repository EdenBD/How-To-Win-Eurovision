```python
# Data exploration:

year_range = (1975, 2016)
finals_only = True

data = votes_df[(votes_df['Year'] >= year_range[0]) & (votes_df['Year'] <= year_range[1])]
if finals_only:
    data = data[votes_df['(semi-) final'] == 'f']

# To see e.g. average of points from a selected country to other countries
# over the year range:
from_country = 'France'

data = data[data['From country'] == from_country]\
                [['To country', 'Points']].groupby('To country').mean()
data.sort_values('Points')
```
